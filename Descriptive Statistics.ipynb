#### Descrptive Statistics
Descriptive statistics are measures that summarize the important features of data, often with a single a number, Producing a descriptive statistics is the first step after cleaning and preparing your data for analysis. Example **mean**, **mode** and **media**

#### Measures of Center

Measure of center are statistics that a give us a sense of the *middle* of a numeric variable. In other words, centrality measures give you a sense of typical value you would expect to see

```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

*load the data from the data folder, we use the mtcars data set from kernel*

```python
mtcars = pd.read_csv("./data/mtcars.csv")
mtcars.index = mtcars.model

del mtcars["model"]

mtcars.head()

```

*Find the means of all the columns in the data set*

```python
mtcars.mean()
```

*Getting the mean values of the rows we need to supply the axis argument*

```python
mtcars.mean(axis=1)
```

*we can find the median for both rows and columns. The median of a distribution is when 50% of the data lies below it and 50% lies above it. In essence the median splits the data in half*

```python
mtcars.median()
```

*We can get the median from all the rows by supplying the same axis value to be 1* Although the mean and median give us some sense of the center of a distribution, they aren't always the same, The **median** is always giving us the value that is splitting the data into two halves whiles the **mean** is a numeric average. so the extreme values can signicantly affect the mean.

```python

# Create DataFrame with normally distributed data
norm_data = pd.DataFrame(np.random.normal(size=100000))

# plot a density graph of a normal distribution
norm_data.plot(kind="density",
              figsize=(10,10),
              label="Distribution")

# plot the mean using a vertical line
plt.vlines(norm_data.mean(),
          ymin=0,
          ymax=0.4,
          linewidth=3.0,
          color="black",
          label="Mean")

# plot the median using a vertical line
plt.vlines(norm_data.median(),
          ymin=0,
          ymax=0.4,
          linewidth=2.0,
          color="red",
          label="Median")
plt.legend()
plt.title("Normal Distribution")
plt.show()
```

*The figure above shows how a normally distributed data looks like, the mean and median are so close to zero that the red median line lies on top of the black thicker mean line.*



In Skewed distribution the mean tends to pull to the direction the skew whiles the median resist that effect

### Positively Skewed Distribution

```python
# Create DataFrame with normally distributed data
skwewed_data = pd.DataFrame(np.random.exponential(size=100000))

# plot a density graph of a normal distribution
skwewed_data.plot(kind="density",
              figsize=(10,10),
              xlim=(-1, 5),
              label="Distribution")

# plot the mean using a vertical line
plt.vlines(skwewed_data.mean(),
          ymin=0,
          ymax=0.8,
          linewidth=3.0,
          color="black",
          label="Mean")

# plot the median using a vertical line
plt.vlines(skwewed_data.median(),
          ymin=0,
          ymax=0.8,
          linewidth=2.0,
          color="red",
          label="Median")
plt.legend()
plt.title("Positively Skewed Distribution")
plt.show()
```

### Negatively skewed distribution

```python
# Create DataFrame with normally distributed data
skwewed_data = pd.DataFrame(-np.random.exponential(size=100000))

# plot a density graph of a normal distribution
skwewed_data.plot(kind="density",
              figsize=(10,10),
              xlim=(-5, 1),
              label="Distribution")

# plot the mean using a vertical line
plt.vlines(skwewed_data.mean(),
          ymin=0,
          ymax=0.8,
          linewidth=3.0,
          color="black",
          label="Mean")

# plot the median using a vertical line
plt.vlines(skwewed_data.median(),
          ymin=0,
          ymax=0.8,
          linewidth=2.0,
          color="red",
          label="Median")
plt.legend()
plt.title("Negatively Skewed Distribution")
plt.show()
```

*The mean is also heavily influenced by outliers whiles the median also resist this influence*

```python
norm_data = np.random.normal(size=50)
outliers_data = np.random.normal(loc=15, scale=1.0, size=3)

combined_data = pd.DataFrame(np.concatenate((norm_data, outliers_data)))


combined_data.plot(kind="density",
                   figsize=(10,10),
                   xlim=(-5, 20),
                   label="Distribution"
                  )

plt.vlines(combined_data.mean(),
           ymin=0,
           ymax=0.2,
           color="black",
           label="Mean",
           linewidth=5.0
          )

plt.vlines(combined_data.median(),
           ymin=0,
           ymax=0.2,
           color="red",
           label="Mean",
           linewidth=3.0
          )

plt.legend()
plt.title("Distribution With Outliers")
plt.show()
```

Since the median tends to resist the effect of outliers and skeweness, It is known as the **robust** statistics. The median generally gives a better sense of a typical value in a distribution with significant skew or outlier

### Measures of Spread

Measures of spread(dispersion) are statistics that describes how data varies, while measures of center give us an idea of a typical value, measures of spread give us a sense of how much the data diverge from the typical value

One of the simplest measures of spread is the **Range**. Range is the distance between the maximum and minimum observations

```python
max(mtcars["mpg"]) - min(mtcars["mpg"])
```

As noted ealier, the median represents the 50th percentile of a dataset, A summary of several percentiles van be used to represent a variable spread. We can extract the **minimum value (0th percentile), first quartile(25th percentile), median, third quartile(75th percentile) and the maximum value (100th percentile)**

```python
five_num_quartile = [mtcars["mpg"].quantile(0),
                    mtcars["mpg"].quantile(0.25),
                    mtcars["mpg"].quantile(0.5),
                    mtcars["mpg"].quantile(0.75),
                    mtcars["mpg"].quantile(1)]

five_num_quartile
```

Since these five values are so commonly used to describe data, they are known as the five number summary. They are the same values that returned from the df.describe()

```python
mtcars["mpg"].describe()
```

#### Outliers Using The Inter-Quartile-Range

The Interquartile range is another common measure of spread, IQR is the difference between the **3rd Quartile and the 1st Quartile**

```python
IQR = mtcars["mpg"].quantile(0.75) - mtcars["mpg"].quantile(0.25)

IQR
```

*To find the outlier we multiply the IQR by 1.5 and subtract from Q1 and also add to Q3*

```python
iqr = ( mtcars["mpg"].quantile(0.25) - (IQR*1.5), mtcars["mpg"].quantile(0.75) + (IQR*1.5))

iqr
```

We find the car model which is an outlier in the mtcars dataset.

```python
lower_bound = mtcars["mpg"].quantile(0.25) - (IQR*1.5)
upper_bound = mtcars["mpg"].quantile(0.75) + (IQR*1.5)


Outliers = mtcars["mpg"][(mtcars["mpg"] < lower_bound) | (mtcars["mpg"] > upper_bound)]
Outliers
```

From the results above it seems the **Toyota Car Model** is the Outlier in the MPG column

We can use the **Box Plot** as the visual representation of the for the 5 number representation and also for **Outliers**

```python
mtcars.boxplot(column="mpg",
              return_type="axes",
              figsize=(8,8))

plt.text(x=0.74, y=22.25, s="3rd Quartile")
plt.text(x=0.8, y=18.75, s="Median")
plt.text(x=0.75, y=15.5, s="1st Quartile")
plt.text(x=0.9, y=10, s="Min")
plt.text(x=0.9, y=33.5, s="Max")
plt.text(x=0.7, y=19.5, s="IQR", rotation=90, size=25)
```

### Measures of Spread

#### Variance and Standard Deviation

The variance and standard deviation are two other common measures of spread. The variance of a distribution is the average of the square deviations from the mean

```python
mtcars["mpg"].var()
```

Standard Deviation is the square root of the variance. Standard deviation can be a bit more interpretable than the variance, since the standard deviation in terms of the same unit as the variable in question, whiles the variance is expressed in terms of units squared

```python
mtcars["mpg"].std()
```

Both variance and standard deviation are derived from the mean, they are susceptible to the influence of data skeweness and outliers. ***Median absolute deviation*** is an alternate measure of spread based in the median, which inherits the median robustness against the influence of skew and outliers. It is the median of the abosulute value of the deviation from the median

```python
abs_median_devs = abs(mtcars["mpg"].median() - mtcars["mpg"])

abs_median_devs.median()
```



### Skeweness and Kurtosis

Beyond the measure of center and spread, descriptive statistics includes measures that give you are sense of the shaoe of the distribution. **Skewness measures the skew or symmetry of a distribution while kurtosis measure how much data is in the tails of a distribution v.s the center**. 

```python
mtcars["mpg"].skew()
```

```python
mtcars["mpg"].kurt()
```

To explore these two mesaures further, let's create some dummy data and inspect it

```python
norm_data = np.random.normal(size=100000)

skewed_data = np.concatenate((np.random.normal(size=35000) + 2,
                             np.random.exponential(size=65000)),
                            axis=0)

uniform_data = np.random.uniform(low=0.2, size=100000)
peaked_data = np.concatenate((np.random.exponential(size=50000),
                             np.random.exponential(size=50000)*(-1)),
                            axis=0)

data_df = pd.DataFrame({
    "norm": norm_data,
    "skewed": skewed_data,
    "uniform": uniform_data,
    "peaked": peaked_data
})
```

```python
data_df.plot(kind="density",
             figsize = (10,10),
             xlim=(-5, 5)
            )
```

Now let's check the skeweness of each of the distribution. Since skeweness measures asymmetry, we'd expect to see low skeweness for all of the distribution except the skewed one, because all the ouliers are roughly symmetric

```python
data_df.skew()
```

Now let's check the kurtosis

```python
data_df.kurt()
```

As we can see from the output, the normally distributed data has a kurtosis near zero, the flat distribution has a negative kurtosis and the two distribution with more data in the tails vs the center have higher kurtosis
